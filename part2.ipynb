```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats # For built-in PDF/PMF functions
```

```python
data2 = np.load('sample_2.npy')

print("Sample 2 size:", data2.shape)
print("Sample 2 min:", data2.min())
print("Sample 2 max:", data2.max())
print("Sample 2 mean:", data2.mean())
print("Sample 2 std:", data2.std())
print("Unique values in Sample 2:", np.unique(data2)[:20])
```

```python
plt.hist(data2, bins=30, density=True, alpha=0.6, edgecolor='black')
plt.title("Histogram of Sample 2")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(data2, bins=30, density=True, alpha=0.6, edgecolor='black', label="Sample 2")


x_vals = np.linspace(0, data2.max(), 200)

scale_est = data2.mean()  # ~ 0.001
rate_est = 1 / scale_est     # ~ 1000
pdf_vals = rate_est * np.exp(-rate_est * x_vals)


plt.plot(x_vals, pdf_vals, 'r-', linewidth=2, label=f"Exponential(λ={rate_est:.1f})")
plt.title("Histogram of Sample 2 + Fitted Exponential PDF")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
