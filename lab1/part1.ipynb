```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats # For built-in PDF/PMF functions
```

```python
sample1 = np.load('sample_1.npy')

print("Sample 1 size:", sample1.shape)
print("Sample 1 min:", sample1.min())
print("Sample 1 max:", sample1.max())
print("Sample 1 mean:", sample1.mean())
print("Sample 1 std:", sample1.std())
print("Unique values in Sample 1:", np.unique(sample1)[:20])
```

```python
for b in [10, 30, 50]:
    plt.hist(sample1, bins=b, density=True, alpha=0.6, edgecolor='black')
    plt.title("Histogram of Sample 1")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

theor_sample = np.random.uniform(50, 51, size=1_000)

plt.figure(figsize=(6, 4))

plt.hist(sample1,
         bins=10,
         density=True,
         alpha=0.5,
         edgecolor='black',
         label="Observed sample1")

plt.hist(theor_sample,
         bins=10,
         density=True,
         alpha=0.5,
         edgecolor='black',
         label=f"Theoretical Uniform")

plt.title("Overlay: Empirical vs. Synthetic Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

* The values lie between 50.0 and 51.0, and the histogram is roughly flat across that interval.
* The sample mean is  50.5, which is the midpoint of 50 and 51.
* The standard deviation is  0.29, which is close to (51 - 50) / sqrt(12)0.289 expected for a Uniform([50,51]).
* Overlying the Synthetic Uniform Distribution shows a very close match.

Thus, Sample 1 is well-approximated by a Uniform distribution on [50,51].
