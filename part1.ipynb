```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats # For built-in PDF/PMF functions
```

```python
sample1 = np.load('sample_1.npy')

print("Sample 1 size:", sample1.shape)
print("Sample 1 min:", sample1.min())
print("Sample 1 max:", sample1.max())
print("Sample 1 mean:", sample1.mean())
print("Sample 1 std:", sample1.std())
print("Unique values in Sample 1:", np.unique(sample1)[:20])
```

```python
plt.hist(sample1, bins=30, density=True, alpha=0.6, edgecolor='black')
plt.title("Histogram of Sample 1")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.hist(sample1, bins=30, density=True, alpha=0.5, edgecolor='black', label="Sample 1")

# The approximate PDF is 1.0 in [50, 51]
x = np.linspace(50, 51, 200)
y = np.ones_like(x)  # all 1s
plt.plot(x, y, 'r-', linewidth=2, label='Uniform(50,51) PDF')

plt.title("Histogram of Sample 1 + Theoretical Uniform PDF")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
